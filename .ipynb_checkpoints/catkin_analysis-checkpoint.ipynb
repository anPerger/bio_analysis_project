```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
pd.set_option("max_columns", None)
```

```python
catkin = pd.read_csv("raw_data/WildlifeCamera-Catkin_Marsh.csv")
catkin
```

```python
catkin = catkin.drop(columns=["Doublecheck_Observers", "Notes", "First_Observer(s)", "File_name"])
catkin["Activity(ies)"] = catkin["Activity(ies)"].fillna("none")
catkin
```

```python
catkin = catkin.dropna()
catkin
```

```python
catkin["Wood_Duck_all"] = catkin["Wood_Duck-M"] + catkin["Wood_Duck-F"] + catkin["Wood_Duck-Baby"]
catkin["Mallard_all"] = catkin["Mallard-M"] + catkin["Mallard-F"] + catkin["Mallard-Baby"]
catkin["Gadwall_all"] = catkin["Gadwall-M"] + catkin["Gadwall-F"] 
catkin
```

```python
catkin["Time (24 hour clock)"] = pd.to_datetime(catkin["Time"])
catkin["Time (24 hour clock)"] = catkin["Time (24 hour clock)"].dt.hour
catkin
```

```python
time_plot_df = catkin[["Time (24 hour clock)", "Wood_Duck_all", "Mallard_all", "Gadwall_all", "Turtle", "Raccoon"]]
sum_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").sum()
sum_time_plot_df
```

```python
ax = sum_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Sum number of each animal seen by hour of the day")
ax.set_ylabel("Sum # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/sum_animals_by_hour.png")
```

```python
mean_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").mean()
mean_time_plot_df
```

```python
ax = mean_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Mean number of each animal seen by hour of the day")
ax.set_ylabel("Mean # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/mean_animals_by_hour.png")
```

```python
max_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").max()
max_time_plot_df
```

```python
ax = max_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Max number of each animal seen by hour of the day")
ax.set_ylabel("Max # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/max_animals_by_hour.png")
```
