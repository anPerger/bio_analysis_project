```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
pd.set_option("max_columns", None)
```

```python
whitaker = pd.read_csv("raw_data/WildlifeCamera-Whitaker_Pond.csv")
whitaker
```

```python
whitaker = whitaker.drop(columns=["Notes", "First_Observers", "Second_Observers", "File_Name"])
whitaker["Activity(ies)"] = whitaker["Activity(ies)"].fillna("none")
whitaker
```

```python
whitaker = whitaker.dropna()
whitaker
```

```python
whitaker["Time (24 hour clock)"] = pd.to_datetime(whitaker["Time"])
whitaker["Time (24 hour clock)"] = whitaker["Time (24 hour clock)"].dt.hour
whitaker
```

```python
time_plot_df = whitaker[["Time (24 hour clock)", "Human", "Coyote", "Raccoon", "Bunny", "Cat", "Skunk"]]
sum_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").sum()
sum_time_plot_df
```

```python
ax = sum_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Sum number of each animal seen by hour of the day")
ax.set_ylabel("Sum # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/whitaker_sum_animals_by_hour.png")
```

```python
mean_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").mean()
mean_time_plot_df
```

```python
ax = mean_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Mean number of each animal seen by hour of the day")
ax.set_ylabel("Mean # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/whitaker_mean_animals_by_hour.png")
```

```python
max_time_plot_df = time_plot_df.groupby("Time (24 hour clock)").max()
max_time_plot_df
```

```python
ax = max_time_plot_df.plot.bar(rot=0, figsize=(20, 16), title="Max number of each animal seen by hour of the day")
ax.set_ylabel("Max # of animals seen")
```

```python
ax = ax.get_figure()
ax.savefig("images/whitaker_max_animals_by_hour.png")
```
